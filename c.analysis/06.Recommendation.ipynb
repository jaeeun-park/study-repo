```python
def recommendation_user(user_id, k):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index # 나하고 유사한 애들 뽑기
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    for i in s:
        # 안 봤던 영화 목록 뽑아내기
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        temp = pd.concat([temp,data[(data.item_id.isin(t)) & (data.rating == 5) & (data.user_id == i)]])
    return temp.drop_duplicates(keep=False).sort_values('item_id')
```

```python
def recommendation_user(user_id, k):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index # 나하고 유사한 애들 뽑기
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    for j,i in enumerate(s, 1):
        # 안 봤던 영화 목록 뽑아내기
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating == 5) & (data.user_id == i)]
        k.rating = k.rating * (1-0.05*j) # 가중치를 나랑 더 유사한 사람은 높이 준다
        temp = pd.concat([temp, k])
        # merge 할 때 이름이 같지 않아 합쳐지지 않으므로 컬럼 명을 바꿔준다.
    temp = temp.rename(columns={'item_id':'movie_id'})
    return temp
```

아예 합쳐서 리턴하자

```python
def recommendation_user(user_id, k):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index # 나하고 유사한 애들 뽑기
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    for j,i in enumerate(s, 1):
        # 안 봤던 영화 목록 뽑아내기
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating == 5) & (data.user_id == i)]
        k.rating = k.rating * (1-0.05*j) # 가중치를 나랑 더 유사한 사람은 높이 준다
        temp = pd.concat([temp, k])
    # merge 할 때 이름이 같지 않아 합쳐지지 않으므로 컬럼 명을 바꿔준다.
    temp = temp.rename(columns={'item_id':'movie_id'})
    # 관련된 정보로 inner join 해준다.
    return temp.merge(item[['movie_id', 'movie_title']]).drop_duplicates()
```

```python
def recommendation_user(user_id, k, n=10):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index # 나하고 유사한 애들 뽑기
    temp = pd.DataFrame(columns=['user_id','item_id','rating'])
    for j,i in enumerate(s, 1):
        # 안 봤던 영화 목록 뽑아내기
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating == 5) & (data.user_id == i)]
        k.rating = k.rating * (1-0.05*j) # 가중치를 나랑 더 유사한 사람은 높이 준다
        temp = pd.concat([temp, k])
    # merge 할 때 이름이 같지 않아 합쳐지지 않으므로 컬럼 명을 바꿔준다.
    temp = temp.rename(columns={'item_id':'movie_id'})
    # 관련된 정보로 inner join 해준다. + 영화 추천수?로 DESC 해서 영화 이름을 반환
    return temp.merge(item[['movie_id', 'movie_title']]).drop_duplicates().groupby('movie_title').mean().sort_values('rating', ascending=False)[:n].index
```
